```python
import seaborn as sns
```

* Lets visualize the distrubution of our features by gender using sns.
* for that we will have to drop CustomerID as it is not relevant to our distrubtion at this this.

```python
df_noID = df.drop(['CustomerID'], axis = 1)
df_noGender = df.drop(['Gender'], axis = 1)
```

```python
sns.pairplot(data=df_noID, hue='Gender')
```

* Lets make a correlation heatmap for the features

```python
mask = np.triu(np.ones_like(df[['Gender','Age','AnnualIncome','Score']].corr(), dtype=bool))
cmap = sns.diverging_palette(230, 20, as_cmap=True)
```

```python
sns.heatmap(df_noID[['Gender','Age','AnnualIncome','Score']].corr(), mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

* Histograms for the all features

```python
for i in df_noID.columns:
    col_num = df_noID[i].value_counts()[:20]
    print("graph for %s: total = %d" % (i, len(col_num)))
    chart = sns.barplot(x=col_num.index, y=col_num)
    chart.set_xticklabels(chart.get_xticklabels(), rotation=90)
    plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler # Visualizing the elbow for number of clusters
```

```python
clusters  = []
```

```python
mms = MinMaxScaler()
mms.fit(df_noGender)
df_noGender_transformed = mms.transform(df_noGender)

Sum_of_squared_distances = []
K = range(1,15)
for k in K:
    km = KMeans(n_clusters=k)
    km = km.fit(df_noGender_transformed)
    Sum_of_squared_distances.append(km.inertia_)
    
plt.plot(K, Sum_of_squared_distances, 'bx-')
plt.xlabel('k')
plt.ylabel('Sum_of_squared_distances')
plt.title('Elbow Method For Optimal k')
plt.show()
```

* It appears that nclusters = 5 would be good enough however we can also make a valid case for nclusters = 3.
* I will do both to compore the results visually.

* Kmeans with n_clusters = 3

```python
kmeans3 = KMeans(n_clusters=3, random_state = 0).fit(df_noGender)

df_noGender['Clusters'] = kmeans3.labels_

#lets visualize our plot
plt.figure(figsize=(12, 8))
sns.scatterplot(df_noGender['AnnualIncome'], df_noGender['Score'], hue=df_noGender['Clusters'], 
                palette=sns.color_palette('hls', 3))
plt.title('KMeans with 3 Clusters')
plt.show()
```

```python
# Kmeans with nclusters = 5
```

```python
kmeans5 = KMeans(n_clusters=5, random_state = 0).fit(df_noGender)

df_noGender['Clusters'] = kmeans5.labels_

#lets visualize our plot
plt.figure(figsize=(12, 8))
sns.scatterplot(df_noGender['AnnualIncome'], df_noGender['Score'], hue=df_noGender['Clusters'], 
                palette=sns.color_palette('hls', 5))
plt.title('KMeans with 5 Clusters')
plt.show()
```
